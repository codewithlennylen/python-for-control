## Plotting Trignonometric Functions




```python
import math
import matplotlib.pyplot as plt

xdata = []
ydata = []

for x in range (0, 25):
    xdata.append(x)
    y = math.sin(x)
    ydata.append(y)

plt.plot(xdata, ydata)
plt.show()
```

### Trigonometric Functions using NumPy

The problem with trigonometric functions from the Math library is that they don't handle arrays as inputs



```python
import numpy as np
import matplotlib.pyplot as plt

xstart = 0
xstop = 2 * np.pi
increment = 0.1

x = np.arange(xstart, xstop, increment)
y = np.sin(x)

plt.plot(x,y)
plt.title("y=sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.axis([0, xstop, -1, 1])
plt.show()


y = np.cos(x)
plt.plot(x,y)
plt.title("y=cos(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.axis([0, xstop, -1, 1])
plt.show()

y = np.tan(x)
plt.plot(x,y)
plt.title("y=tan(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.axis([0, xstop, -1, 1])
plt.show()
```
