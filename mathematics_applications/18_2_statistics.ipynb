## 18.2.1. Introduction to Statistics

### Mean or average:

The mean is the sum of the data divided by the number of data points.



### Variance

Variance is the measure of variation in a dataset



### Standard Deviation

The standard deviation is a measure of the spread of the values in a dataset or the value of a random variable.
It is defined as the square root of the variance.



```python
import statistics as st
import numpy as np # You can also use NumPy

data = [1,2.5,3,4,5,6.7]

m = st.mean(data)
print(f"mean: {m}")

med = st.median(data)
print(f"median: {med}")

var = st.variance(data)
print(f"variance: {var}")

std = st.stdev(data)
print(f"Standard Deviation: {std}")

std = np.std(data)
print(f"Standard Deviation: {std}")

minimum = np.min(data)
print(f"minimum data: {minimum}")

maximum = np.max(data)
print(f"maximum data: {maximum}")
```
