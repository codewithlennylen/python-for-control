```python
from math import asin, atan, acos, sqrt
import mymath


def findA(i,j,k):

    if k == 'ac':
        a = asin(i/j)
        A = mymath.r2d(a)
    
    elif k == "ab":

        a = atan(i/j)
        A = mymath.r2d(a)
    
    elif k == 'bc':

        a = acos(i/j)
        A = mymath.r2d(a)

    return A

a = 5
b = 8
c = sqrt((a**2)+(b**2))
print(c)

print(findA(a,b,"ab"))
print(findA(a,c,"ac"))
print(findA(b,c,"bc"))
```

```python
from math import cos, sqrt, pi

def findC(a,b, C):
    c_squared = (a**2) + (b**2) - (2*(a*b)*cos(C))

    c = sqrt(c_squared)

    return c

print(findC(a,b, pi/2))
```
