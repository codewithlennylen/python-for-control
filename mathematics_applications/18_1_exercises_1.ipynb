```python
from math import log, sqrt, e

def z(x, y):
    a = 3 * (x ** 2) # (x * x)
    b = sqrt((x ** 2)+(y ** 2))
    c = pow(e, log(x)) # e ** log(x)

    return a + b + c

print(z(2,3))
print(z(1,1))
```

```python
from math import sin, pi, cos

def f(a,b,c, x):
    bx_c = (b * x) + c
    ax2 = (a * (x ** 2))
    a = log(ax2 + bx_c)
    b = sin(ax2 + bx_c)
    numerator = a - b

    c = (4 * pi * (x ** 2))
    d = cos(x - 2) * (ax2 + bx_c)
    denominator = c + d

    return numerator / denominator

print(f(1,3,5, 9))
```
